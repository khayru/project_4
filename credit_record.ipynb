```python
# Remove unnecessary columns
credit_application_clean= credit_application.drop(columns=['FLAG_MOBIL','FLAG_WORK_PHONE','FLAG_PHONE','FLAG_EMAIL'])
credit_application_clean
```

```python
# Assign the category and subcategory values to category and subcategory columns.
credit_application_clean[['Married','Single ']] = credit_application_clean ["NAME_FAMILY_STATUS"].str.split('/' , expand=True)
credit_application_clean
```

```python
# Replace 'C' and 'X' values with 0
credit_record['STATUS'].replace({'C': 0, 'X': 0}, inplace=True)

# Convert the 'STATUS' column to integers
credit_record['STATUS'] = credit_record['STATUS'].astype(int)
```

```python
average_status_by_customer = credit_record.groupby('ID')['STATUS'].mean().reset_index()

# Rename the column for clarity
average_status_by_customer.rename(columns={'STATUS': 'Average_Status'}, inplace=True)

# Display the resulting DataFrame
average_status_by_customer['Average_Status'].sort_values()
```

```python
average_status_by_customer['Average_Days_Late'] = average_status_by_customer['Average_Status'] * 30

# Display the resulting DataFrame with average days late
print(average_status_by_customer)
```

```python
credit_application_clean = credit_application_clean.merge(average_status_by_customer[['ID', 'Average_Status']], on = 'ID', how = 'left')
credit_application_clean
```
